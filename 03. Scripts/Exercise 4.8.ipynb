# Importing Libraries and Data

```python
# Libraries
import pandas as pd
import numpy as np
import os
```

```python
# data path
path = r'/Users/Evan/Desktop/Instacart Analysis'
```

```python
# Data
df_opm = pd.read_pickle(os.path.join(path, '02. Data', 'Prepared Data', 'orders_products_merged_frequencies.pkl'))
```

# Step 2. 

```python
# avg orders by department
df_opm.groupby('department_id').agg({'order_number':['mean']})
```

Most departments hold a higher average when weighted against the entire dataframe, however departments 1, 2, 6, 14, 16, 18, and 21 all hold higher averages in the subset. 

# Step 4. 

```python
# creating max order column
df_opm ['max_order'] = df_opm.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# loayl customer flag
df_opm.loc[df_opm['max_order'] > 40, 'loyalty_flag'] = 'loyal customer'
```

```python
# regular customer flag
df_opm.loc[(df_opm['max_order'] <= 40) & (df_opm ['max_order'] >= 10), 'loyalty_flag'] = 'regular customer'
```

```python
# new customer flag
df_opm.loc[df_opm['max_order'] < 10, 'loyalty_flag'] = 'new customer'
```

```python
# row limit removal
pd.options.display.max_rows = None
```

```python
# flag check
df_opm.head(100)
```

# Step 5

```python
# average price by loyalty flag
df_opm.groupby('loyalty_flag').agg({'prices' : ['mean']})
```

# Step 6. 

```python
# Creating avg item spend column
df_opm ['avg_item_spend'] = df_opm.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# data check
df_opm.head()
```

```python
# high spender flag
df_opm.loc[df_opm['avg_item_spend'] >= 10, 'spend_flag'] = 'high spender'
```

```python
# low spender flag
df_opm.loc[df_opm['avg_item_spend'] < 10, 'spend_flag'] = 'low spender'
```

```python
# flag check
df_opm.head()
```

# Step 7

```python
# creating med days since prior order column
df_opm ['med_days_since_prior_order'] = df_opm.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# NF flag
df_opm.loc[df_opm['med_days_since_prior_order'] > 20, 'freq_flag'] = 'non-frequent customer'
```

```python
# reg flag
df_opm.loc[(df_opm['med_days_since_prior_order'] <= 20) & (df_opm ['med_days_since_prior_order'] > 10), 'freq_flag'] = 'regular customer'
```

```python
# frequent flag
df_opm.loc[df_opm['med_days_since_prior_order'] <= 10, 'freq_flag'] = 'frequent customer'
```

```python
# data check
df_opm.head(50)
```

```python
df_opm.loc[df_opm['prices'] > 100] = np.nan
```

```python
df_opm.drop(columns = ['Unnamed: 0.1', 'Unnamed: 0'])
```
