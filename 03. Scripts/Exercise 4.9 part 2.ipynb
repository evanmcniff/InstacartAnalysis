# Importing Libraries and Data

```python
# Libraries
import pandas as pd
import numpy as np 
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
# File Path
path = r'/Users/Evan/Desktop/Instacart Analysis'
```

```python
# Data
df = pd.read_pickle(os.path.join(path, '02. Data', 'Prepared Data', 'ord_prod_cust_combined.pkl'))
```

# Bar Chart

```python
# bar chart
bar = df['order_day_of_week'].value_counts().sort_index().plot.bar(title = 'Order Frequency by Day', xlabel = 'Day of Week', ylabel = 'Frequency')
```

# Histogram

```python
# histogram value limit
df.loc[df['prices'] > 100] = np.nan
```

```python
# Histogram
hist = df['prices'].plot.hist(bins = 25, title = 'Price Distribution')
```

# Scatterplot

```python
# Scatterplot (bound by histogram value limits)
scat = sns.scatterplot(x = 'prices', y = 'prices', data = df)
```

# Line Chart

```python
# Random seed to generate 70/30 subset 
np.random.seed(4)
dev = np.random.rand(len(df))<= 0.7
```

```python
# dev check
dev
```

```python
# > 0.7
big = df[dev]
```

```python
# < 0.7
small = df[~dev]
```

```python
# data check
len(df)
```

```python
# data check
len(big) + len(small)
```

```python
df_2 = small[['order_day_of_week', 'prices']]
```

```python
# Line chart
line = sns.lineplot(data = df_2, x = 'order_day_of_week', y = 'prices')
```

# Step 3

```python
# Histogram for houlry order frequency
hist2 = df['order_hour_of_day'].plot.hist(bins = 24, title = 'Orders by Hour')
```

The historgram demonstrates that the busiest period of the day occurs between hours 10 and 15. Most customers are ordering in the late morning and early afternoon hours of the day. 

# Step 4

```python
# loyalty bar chart
bar2 = df['loyalty_flag'].value_counts().sort_index().plot.bar(title = 'Loyalty Distribution')
```

# Step 6

```python
# age/dependants line chart
line3 = sns.lineplot(data = df_4, x = 'Age', y = 'dependants')
```

# Step 7

```python
# age/income scatteplot
scat2 = sns.scatterplot(x = 'Age', y = 'income', data = df)
```

The scatterplot here shows us that income from age 20-40 is fairly consistent with some earners exceeding 200k mark, but from age 40 and beyond there are far more customers above the 200k mark and even some that exceed the 400k mark which is the peak earning level for customer age 20-40.
