# User Defined Functions

```python
# user defined function to denote price ranges
def price_label(row):
    if row ['prices'] <=5:
        return 'low-range product'
    elif row ['prices'] >5 and row ['prices'] <=15:
        return 'mid-range product'
    elif row['prices'] >15:
        return 'high-range product'
    else: return 'not enough data'
```

```python
# apply function
df['price_label'] = df.apply (price_label, axis = 1)
```

```python
# check function counts
df['price_label'].value_counts()
```

```python
df['prices'].max()
```

# For Loops

```python
# daily order counts
df_opc['order_day_of_week'].value_counts(dropna = False)
```

```python
# defining days of the week
result = []
for value in df_opc['order_day_of_week']:
    if value == 0:
        result.append('busiest day')
    elif value == 4:
        result.append('least busy')
    else:
        result.append ('regularly busy')
```

```python
result
```

```python
#creating columnn
df_opc['busiest day'] = result
```

```python
# new column counts
df_opc['busiest day'].value_counts(dropna = False)
```

# Task

```python
# defining busiest days
result = []
for value in df_opc['order_day_of_week']:
    if value == (0 or 1):
        result.append('busiest days')
    elif value == (3 or 4):
        result.append('least busy days')
    else:
        result.append ('regularly busy')
```

```python
# checking result
result
```

```python
# creating column
df_opc['busiest days'] = result
```

```python
# value check
df_opc['busiest days'].value_counts(dropna = False)
```

busiest days are Saturday and Sunday which make sense as most people have the time and ability to order on weekends. While the least busy days are Tuesday and Wednesday. Nearly 2m more orders are placed on the weekend days than on the slowest weekdays. 

```python
# checking hourly counts
df_opc['order_hour_of_day'].value_counts(dropna = False)
```

```python
# defining busiest hours
result = []
for value in df_opc['order_hour_of_day']:
    if value == (10 or 11 or 14 or 15 or 13 or 12 or 16 or 9):
        result.append('most orders')
    elif value == (3 or 4 or 2 or 5 or 1 or 0 or 6 or 23):
        result.append('least orders')
    else:
        result.append ('average orders')
```

```python
# creating new column
df_opc['busiest_period_of_day'] = result
```

```python
# Checking column counts
df_opc['busiest_period_of_day'].value_counts(dropna = False)
```
